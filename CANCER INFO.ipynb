```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the cancer quality dataset
cancer_dataset = pd.read_csv("C:/Users/USER/Desktop/New folder/cancer/cancer patient data sets.csv")
```

```python
cancer_dataset.shape
```

```python
# Print the first 5 rows of the dataset
cancer_dataset.head()
```

```python
# check if there are null values in the datasets
cancer_dataset.isnull().sum()
```

```python
# Scatter plot to show correlation between chronic lung disease and genetic risk of cancer patients
plt.figure(figsize=(12, 7))
sns.regplot(x='chronic Lung Disease', y='Genetic Risk', data=cancer_dataset, scatter_kws={'s': 100}, line_kws={'color': 'red'}, ci=None)
plt.title('Correlation between chronic Lung Disease and Genetic Risk ')
plt.xlabel('chronic Lung Disease')
plt.ylabel('Genetic Risk')
plt.grid(True)
plt.show()
```

```python
#Create a histogram showing age distribution of cancer patients
sns.set(style="whitegrid")  # Set the plot style
plt.figure(figsize=(8, 6))  # Set the figure size

# Plot the histogram
sns.histplot(data=cancer_dataset, x='Age', bins=10, kde=True, color='blue')

# Add labels and a title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Cancer Patients')

# Show the plot
plt.show()
```

```python
# Create a bar chart to depict gender distribution
sns.set(style="whitegrid")  # Set the plot style
plt.figure(figsize=(6, 4))  # Set the figure size

# Plot the bar chart
sns.countplot(data=cancer_dataset, x='Gender', palette="Set3")

# Add labels and a title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Gender Among Cancer Patients')

# Show the plot
plt.show()
```

```python
# Design a correlation Heatmap of Lung Health Variables for cancer patients
# Select the relevant columns for lung health
lung_health_data = cancer_dataset[['chronic Lung Disease', 'Shortness of Breath', 'Wheezing']]

# Calculate the correlation matrix
correlation_matrix = lung_health_data.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add labels and a title
plt.xlabel('Lung Health Variables')
plt.ylabel('Lung Health Variables')
plt.title('Correlation Heatmap of Lung Health Variables')

# Show the plot
plt.show()
```
