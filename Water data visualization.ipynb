```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read the water quality dataset
water_dataset = pd.read_csv("C:/Users/USER/Desktop/New folder/Water Quality and Potability/water_potability.csv")

# Print the first 5 rows of the dataset
water_dataset.head()
```

```python
# check if there are duplicates
water_dataset.duplicated().sum()
```

```python
# check if there are null values in the datasets
water_dataset.isnull().sum()
```

```python
# give a full description of your dataset
water_dataset.describe()
```

```python
#visualise the data distribution of potability
plt.figure(figsize=(5,5))
sns.countplot(data=water_dataset, x=water_dataset["Potability"], color="blue")
plt.show()
```

```python
# Create a scatter plot of the pH and Conductivity columns of the dataset
water_dataset.plot(x="ph", y="Conductivity", kind="scatter")
```

```python
# Generate a heatmap of the correlation matrix of the dataset
plt.figure(figsize=(7,5))
sns.heatmap(water_dataset.corr(), annot=True, cmap="Greens", fmt=".2f")
plt.show()
```

```python
# Lets create a scatter plot for the distribution of the numeric column with respect to ph
columns=water_dataset.select_dtypes(exclude="object").columns
for i in range(len(columns)-1):
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=water_dataset, x=water_dataset['ph'], y=water_dataset[columns[i]], color="blue")
    plt.title(f"{columns[i]} vs ph")
    plt.show()
```

```python
# visualise the distribution of the numerical value with respect to Potability
columns=water_dataset.select_dtypes(exclude="object").columns
for i in range(len(columns)-1):
    plt.figure(figsize=(8,5))
    sns.histplot(data=water_dataset, x=water_dataset[columns[i]], hue="Potability", kde=True, color="red", multiple="stack", bins=15)
    plt.title(f"{columns[i]} vs Potability")
    plt.show()
```
